## 원주 기온데이터 분석

```python
import pandas as pd
import numpy as np
import csv
```

### 공공데이터를 데이터프레임으로 읽어오기

```python
f = open(r'wonju_temp.csv','r', encoding='euc-kr')
```

```python
data = csv.reader(f, delimiter=',')
head = next(data)
```

```python
df = pd.DataFrame(data)
```

```python
f.close()
```

```python
df.columns=['지역코드','지역','일시','평균기온','최고기온','최고기온시각','최저기온','최저기온시각','일교차']
```

### 데이터 탐색 및 전처리

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df['일시'].replace('',np.nan, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df.astype({'평균기온':'float','최고기온':'float','최저기온':'float'})
```

```python
df.info()
```

```python
df_wonju = df[['일시','평균기온','최고기온','최저기온']]
```

```python
df_wonju
```

### 데이터 분석 1 : 가장 더웠던 기온은 몇도이며 언제였나

```python
high_temp = -999

for a, b in zip(df_wonju['최고기온'], df_wonju['일시']):
    if a > high_temp:
        high_temp = a
        high_date = b
```

```python
print('날짜 =>', high_temp, '\t기온 =>', high_date)
```

## 데이터 분석 2 : 가장 추웠던 기온은 몇도이며 언제였나

```python
low_temp = 999

for a, b in zip(df_wonju['최고기온'], df_wonju['일시']):
    if a < low_temp:
        low_temp = a
        low_date = b
```

```python
print('날짜 =>', low_date, '\t기온 =>', low_temp)
```

### 시각화 (Matplotlib)

```python
import warnings
warnings.filterwarnings('ignore')
from matplotlib import pyplot as plt
%matplotlib inline
import platform

if platform.system() == 'Windows':
    plt.rc('font',family='Malgun Gothic')
    plt.rcParams['font.size']=10
```

```python
label = ['최고기온','최저기온','평균기온']

plt.plot(df_wonju['최고기온'], label='최고기온', color='red', linestyle='-')#, marker='o')
plt.plot(df_wonju['최저기온'], label='최저기온', color='blue', linestyle='-')#, marker='s')
plt.plot(df_wonju['평균기온'], label='평균기온', color='#FFDB01', linestyle='-')#, marker='v')

# 상단 좌측
plt.legend(loc='upper left')

plt.title('지난 10년간의 2월달 온도 변화')
plt.ylabel('기온')

plt.ylim(-20,50)
plt.grid(False)
plt.rcParams['figure.figsize']=(8,3)
plt.show()
```
