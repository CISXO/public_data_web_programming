```python
wonju_cafe = df[['storeName','gpsX','gpsY']]
```

```python
wonju_loc = fo.Map(location=[37.28, 127.90], zoom_start=15)
```

```python
wonju_loc
```

```python
for a, b in zip(wonju_cafe['gpsX'], wonju_cafe['gpsY']):
    fo.Marker([a,b]).add_to(wonju_loc)
```
