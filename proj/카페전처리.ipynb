```python
chun_cafe = df[['storeName','gpsX','gpsY']]
chun_loc = fo.Map(location=[37.28, 127.91], zoom_start=14)
chun_loc
```

```python
for a, b in zip(chun_cafe['gpsX'], chun_cafe['gpsY']):
    fo.Marker([a, b]).add_to(chun_loc)
fo.Marker([37.28, 127.91]).add_to(chun_loc)
chun_loc
```

# 뉴스 크롤링

```python
import requests
import html5lib
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://news.naver.com'
html = requests.get(url).text
soup = BeautifulSoup(html, "html5lib")
tags = soup.find_all('div', "cjs_t")
for tag in tags:
    print (tag.text)
```

```python
url = 'https://weather.naver.com'
html = requests.get(url).text
soup = BeautifulSoup(html, "html5lib")
tags = soup.find_all('strong', "current")
for tag in tags:
    print (tag.text)
```
